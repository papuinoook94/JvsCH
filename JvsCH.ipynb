```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
data=pd.read_csv("Big_Japan_vs_China_Technology.csv")
data
```

```python
data.head()
```

```python
data.describe(include='all')
```

```python
# 1. Comparamos el crecimiento de sectores tecnológicos entre China y Japón
# Agrupamos los datos por país y sector tecnológico para obtener la participación de mercado promedio
market_share_country_sector = data.groupby(['Country', 'Tech Sector'])['Market Share (%)'].mean().unstack()
print(market_share_country_sector)
```

```python
# Agrupamos por país y sector
market_share_country_sector = data.groupby(['Country', 'Tech Sector'])['Market Share (%)'].mean().unstack()

# Creamos un gráfico de líneas
plt.figure(figsize=(12, 6))

# Iteramos sobre cada sector tecnológico para agregar líneas al gráfico
for sector in market_share_country_sector.columns:
    plt.plot(market_share_country_sector.index, 
             market_share_country_sector[sector], 
             marker='o', 
             label=sector)

# Añadimos título y etiquetas
plt.title('Comparación del Crecimiento de Sectores Tecnológicos entre China y Japón', fontsize=14)
plt.xlabel('País', fontsize=12)
plt.ylabel('Participación de Mercado Promedio (%)', fontsize=12)
plt.xticks(rotation=0)  # Mantener etiquetas de países horizontales

# Añadimos leyenda
plt.legend(title='Sector Tecnológico')
plt.grid(True)
plt.tight_layout()

# Mostramos el gráfico
plt.show()
```

```python
# 2. Inversión en I+D por país
# Agrupamos por país y calcular la inversión promedio en investigación y desarrollo (I+D)
rd_investment_by_country = data.groupby('Country')['R&D Investment (in USD)'].mean()
print(rd_investment_by_country)
```

```python

# Agrupamos los datos por país, año y sector, y calcular la inversión total en I+D
rd_investment_over_time = data.groupby(['Country', 'Year', 'Tech Sector'])['R&D Investment (in USD)'].sum().reset_index()

# Obtenemos la lista de sectores únicos
sectors = rd_investment_over_time['Tech Sector'].unique()

# Creamos un gráfico para cada sector
for sector in sectors:
    plt.figure(figsize=(10, 6))
    
    # Filtramos los datos del sector actual
    sector_data = rd_investment_over_time[rd_investment_over_time['Tech Sector'] == sector]
    
    # Graficamos los datos de cada país
    for country in ['China', 'Japan']:
        country_data = sector_data[sector_data['Country'] == country]
        plt.plot(country_data['Year'], country_data['R&D Investment (in USD)'], marker='o', label=country)
    
    plt.title(f'Inversión en I+D en el Sector: {sector}')
    plt.xlabel('Año')
    plt.ylabel('Inversión en I+D (USD)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    
    # Mostramos los gráficos
    plt.tight_layout()
    plt.show()
```

```python
# 3. Relación entre la inversión en I+D y el número de patentes
# Calculamos la correlación entre la inversión en I+D y el número de patentes presentadas anualmente
correlation_rd_patents = data[['R&D Investment (in USD)', 'Number of Patents Filed (Annual)']].corr().iloc[0, 1]
print(correlation_rd_patents)
```

```python
# Creamos un gráfico de dispersión para visualizar la relación
plt.figure(figsize=(10, 6))

# Gráfico de dispersión
sns.scatterplot(data=data, x='R&D Investment (in USD)', y='Number of Patents Filed (Annual)', hue='Country', style='Tech Sector', s=100)

# Ajustar una línea de regresión
sns.regplot(data=data, x='R&D Investment (in USD)', y='Number of Patents Filed (Annual)', scatter=False, color='black')

# Personalizar el gráfico
plt.title('Relación entre Inversión en I+D y Número de Patentes Presentadas')
plt.xlabel('Inversión en I+D (USD)')
plt.ylabel('Número de Patentes Presentadas (Anual)')
plt.legend(title='País y Sector Tecnológico')
plt.grid()

# gráfico
plt.tight_layout()
plt.show()
```

```python
# 4. Innovación global - Clasificación promedio por país
# Calculamos el promedio del ranking de innovación global para China y Japón
innovation_ranking_country = data.groupby('Country')['Global Innovation Ranking'].mean()
print(innovation_ranking_country)
```

```python
plt.scatter(data['R&D Investment (in USD)'], data['Global Innovation Ranking'])
plt.title("Relación entre Ranking de Innovación Global e Inversión en I+D")
plt.xlabel("Inversión en I+D (USD)")
plt.ylabel("Ranking de Innovación Global")
plt.show()
```

```python
# 5. Impacto de la cobertura de 5G y la penetración de Internet en Startups y Empresas tecnológicas
# Calculamos la correlación entre la cobertura de 5G y el número de startups
correlation_5g_startups = data[['5G Network Coverage (%)', 'Number of Startups']].corr().iloc[0, 1]
print(correlation_5g_startups)
# Calculamos la correlación entre la penetración de internet y el número de empresas tecnológicas
correlation_internet_companies = data[['Internet Penetration (%)', 'Number of Tech Companies']].corr().iloc[0, 1]
print(correlation_internet_companies)
```

```python
# 6.Agrupamos los datos por país y año, y sumar el número de startups
startups_by_country_year = data.groupby(['Country', 'Year'])['Number of Startups'].sum().unstack()

# Resumen: Mostrará la tendencia de crecimiento de startups a lo largo de los años para China y Japón.
print("Distribución de startups por país y año:")
print(startups_by_country_year)
```

```python
startups_by_country_year = data.groupby(['Country', 'Year'])['Number of Startups'].sum().unstack()

# Crear un gráfico de barras
plt.figure(figsize=(12, 6))

# Crear las barras para cada país
startups_by_country_year.plot(kind='bar', width=0.8)

# Personalizar el gráfico
plt.title('Número de Startups por País y Año')
plt.xlabel('Año')
plt.ylabel('Número de Startups')
plt.xticks(rotation=45)
plt.legend(title='País')
plt.grid(axis='y')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# 7. Calcular la correlación entre el número de startups y la inversión en capital de riesgo
correlation_startups_vc = data[['Number of Startups', 'Venture Capital Funding (in USD)']].corr().iloc[0, 1]
```

```python
# Creamos un gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number of Startups', y='Venture Capital Funding (in USD)', data=data, color='blue')

# Agregar una línea de regresión
sns.regplot(x='Number of Startups', y='Venture Capital Funding (in USD)', data=data, scatter=False, color='orange')

# Configuración de la gráfica
plt.title('Correlación entre el Número de Startups y la Inversión en Capital de Riesgo', fontsize=16)
plt.xlabel('Número de Startups', fontsize=14)
plt.ylabel('Inversión en Capital de Riesgo (USD)', fontsize=14)
plt.grid(True)
plt.show()

# Imprimir la correlación
print("\nCorrelación entre el número de startups y la inversión en capital de riesgo:")
print(correlation_startups_vc)
```

```python
#8. Agrupar por país y contar los productos tecnológicos exportados
top_products_exported = data.groupby('Country')['Top Tech Products Exported'].value_counts().unstack()
```

```python

# Configuración del gráfico
plt.figure(figsize=(12, 8))
top_products_exported.plot(kind='bar', stacked=True, colormap='viridis', ax=plt.gca())

# Configuración adicional
plt.title('Principales Productos Tecnológicos Exportados por País', fontsize=16)
plt.xlabel('País', fontsize=14)
plt.ylabel('Cantidad de Productos Exportados', fontsize=14)
plt.xticks(rotation=0)  # Mantener las etiquetas en horizontal
plt.legend(title='Productos Tecnológicos', bbox_to_anchor=(1.05, 1), loc='upper left')  # Ubicación de la leyenda
plt.tight_layout()  # Ajustar el gráfico para que se vea bien
plt.show()

# Resumen de los productos exportados
print("\nPrincipales productos tecnológicos exportados por país:")
print(top_products_exported)
```

```python
#9. Calcular la correlación entre el número de empresas tecnológicas y el número de trabajadores tecnológicos
correlation_companies_workers = data[['Number of Tech Companies', 'Number of Tech Workers']].corr().iloc[0, 1]

```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number of Tech Companies', y='Number of Tech Workers', data=data)

# Añadir una línea de regresión
sns.regplot(x='Number of Tech Companies', y='Number of Tech Workers', data=data, scatter=False, color='red')

# Configuración adicional
plt.title('Relación entre el Número de Empresas Tecnológicas y el Número de Trabajadores Tecnológicos', fontsize=16)
plt.xlabel('Número de Empresas Tecnológicas', fontsize=14)
plt.ylabel('Número de Trabajadores Tecnológicos', fontsize=14)
plt.xlim(0, data['Number of Tech Companies'].max() + 10)  # Ajustar el límite del eje x
plt.ylim(0, data['Number of Tech Workers'].max() + 10)  # Ajustar el límite del eje y
plt.grid(True)
plt.tight_layout()  # Ajustar el gráfico para que se vea bien
plt.show()

# Resumen de la correlación
print("\nCorrelación entre el número de empresas tecnológicas y trabajadores tecnológicos:")
print(correlation_companies_workers)
```

```python
#10. Calcular la correlación entre el ranking de innovación global y la inversión en I+D
correlation_innovation_rd = data[['Global Innovation Ranking', 'R&D Investment (in USD)']].corr().iloc[0, 1]

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Global Innovation Ranking', y='R&D Investment (in USD)', data=data)

# Añadir una línea de regresión
sns.regplot(x='Global Innovation Ranking', y='R&D Investment (in USD)', data=data, scatter=False, color='red')

# Configuración adicional
plt.title('Relación entre el Ranking de Innovación Global y la Inversión en I+D', fontsize=16)
plt.xlabel('Ranking de Innovación Global', fontsize=14)
plt.ylabel('Inversión en I+D (USD)', fontsize=14)
plt.xlim(data['Global Innovation Ranking'].min() - 1, data['Global Innovation Ranking'].max() + 1)  # Ajustar el límite del eje x
plt.ylim(0, data['R&D Investment (in USD)'].max() + 1000000)  # Ajustar el límite del eje y
plt.grid(True)
plt.tight_layout()  # Ajustar el gráfico para que se vea bien
plt.show()

# Resumen de la correlación
print("\nCorrelación entre el ranking de innovación global y la inversión en I+D:")
print(correlation_innovation_rd)
```
